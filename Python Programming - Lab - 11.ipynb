```python
import matplotlib.pyplot as plt
import matplotlib.image as imgplt
import numpy as np
import random
```

### 01) WAP to demonstrate the use of Pie chart.

```python
values=[1,2,3,4,5,6,7]
labs=["food","movie","daku","Accos","misc","ufo","kng"]
clrs=["r","g","b","c","m","g","c"]
expd=[0 for i in range(1,8)]
expd[6]=0.1
plt.pie(values,labels=labs,colors=clrs,explode=expd)
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
val1=[random.randrange(1,11) for i in range(1,11)]
val2=[random.randrange(1,11) for i in range(1,11)]
```

```python
plt.plot(range(1,11),val1,ls="",c="r",marker="*")
plt.plot(range(1,11),val2,c="b",marker="*",ls="")
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
plt.bar([i for i in range(1,6)],[random.randrange(1,100) for i in range(1,6)],color=["red","blue","red","blue","red"])
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
plt.hist(np.random.randint(0,10,100),bins=10,histtype="stepfilled")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
cag=["car","bike","tuk-tuk"]
sales=[100,200,50]
plt.bar(cag,sales)
for i,value in enumerate(sales):
    plt.text(i,value+1,str(value),ha="center")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?
